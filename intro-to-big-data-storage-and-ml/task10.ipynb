```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('test10.csv')
df
```

```python
mdl = KMeans(n_clusters=3, 
             init=np.array([[12.6, 8.8], [13.0, 12.6], [13.0, 11.8]]), 
             max_iter=100, 
             n_init=1).fit(df[['X', 'Y']])
mdl.labels_
```

```python
mdl.cluster_centers_
```

```python
df['pred'] = mdl.labels_
```

```python
df
```

```python
zero_df = df[df.pred == 0]
zero_df
```

```python
zero_df['dist'] = np.sqrt((zero_df.X - 11.4)**2 + (zero_df.Y - 4.8)**2)
```

```python
zero_df
```

```python
zero_df.dist.mean()
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, auc, roc_auc_score
```

```python
df = pd.read_csv('candy-data.csv')
df
```

```python
test = pd.read_csv('candy-test.csv')
X_test = test[['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 
                   'crispedricewafer', 'hard', 'bar', 'pluribus', 'sugarpercent', 'pricepercent']]
y_test = test['Y']
```

```python
test
```

```python
train_df = df[(df['competitorname'] != 'Mike & Ike') & (df['competitorname'] != 'Root Beer Barrels') \
              & (df['competitorname'] != 'Starburst')]
```

```python
X_train = train_df[['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 
                   'crispedricewafer', 'hard', 'bar', 'pluribus', 'sugarpercent', 'pricepercent']]
y_train = train_df['Y']
```

```python
mdl = LogisticRegression(random_state = 2019, solver = 'lbfgs').fit(X_train, y_train)
```

```python
mdl.predict_proba(X_test)[7]
```

```python
precision_score(y_test, mdl.predict(X_test))
```

```python
recall_score(y_test, mdl.predict(X_test))
```

```python
roc_auc_score(y_test, mdl.predict_proba(X_test)[:, 1])
```
